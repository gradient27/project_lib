### Задание 1

Загрузка библиотек

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
```

Загрузка образцов из датасета бостон

```python
boston = load_boston()
```

```python
boston
```

```python
feature_names = boston["feature_names"]

feature_names
```

```python
target = boston["target"]

target[:10]
```

```python
data = boston["data"]
```

```python
X = pd.DataFrame(data, columns=feature_names)

X.head()
```

```python
X
```

```python
y = pd.DataFrame(target, columns=["price"])
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
})
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

### Задание 2

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(criterion='mse', max_depth=12, #глубина дерева
                                min_samples_leaf=30, #минимальное количество наблюдений в листе дерева
                                random_state = 42,
                                n_estimators=1000 #количество деревьев
                                )

rf_model.fit(X_train, y_train.values[:, 0])
y_train_preds = rf_model.predict(X_train)
r2_score(y_train.values[:, 0], y_train_preds)
```

Данный метод позволяет лучше прогнозировать значения
